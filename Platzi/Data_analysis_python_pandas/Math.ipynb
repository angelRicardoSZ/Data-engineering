```python
def fun_1(x):
  y = x**2 + 1
  return y
```

```python
fun_1(10)
```

```python
np.arange(-5,6)
```

```python
fun_1(np.arange(-5,6))
```

```python
df["hour"].apply(fun_1)
```

```python
def fun_2(x,a=1,b=0):
  y = x**2 + a*x + b
  return y
```

```python
fun_2(10, a=20, b=-100)
```

```python
df["hour"].apply(fun_2, args=(20,-100))
```
