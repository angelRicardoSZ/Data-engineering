```python
df_pdb = pd.read_csv("poblacion.csv")
pd.options.display.float_format = "{:,.1f}".format
df_pdb
```

```python
df_pdb["year"] = pd.Categorical(df_pdb["year"].apply(str))
df_pdb["year"].head()
```

```python
df_pdb.dtypes
```

```python
idx_filtro = df_pdb["Country"].isin(["Aruba", "Colombia"])
idx_filtro
```

```python
df_sample = df_pdb[idx_filtro]
df_sample
```

```python
df_sample = df_sample.set_index(["Country", "year"]).sort_index()
df_sample
```

```python
df_sample.loc["Colombia",:].loc["2016"]
```

```python
df_sample.xs(["Aruba"])
```

```python
df_sample.xs("2018", level="year")
```

```python
df_countries = df_pdb.set_index(["Country", "year"]).sort_index(ascending=[True,True])
df_countries
```
