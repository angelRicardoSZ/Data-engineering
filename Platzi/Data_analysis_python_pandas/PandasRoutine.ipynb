```python
df_temp = pd.read_csv('GlobalLandTemperaturesByCity.csv')
df_temp
```

```python
df_temp["Country"].unique()
```

```python
df_temp["year"] = pd.to_datetime(df_temp["dt"])
df_temp["year"]
```

```python
idx = df_temp['year'] > pd.to_datetime('1970-01-01')
df_temp = df_temp[idx]
df_temp
```

```python
df_temp.groupby(["Country", pd.Grouper(key="year",freq="1Y")]).aggregate({"AverageTemperatureUncertainty":[np.mean,np.median]})
```

```python
df_t_avg = df_temp.groupby(["Country", pd.Grouper(key="year",freq="1Y")]).agg({"AverageTemperature":[np.mean,np.median]})
```

```python
df_t_avg
```

```python
df_t_avg.xs("Mexico")["AverageTemperature"].plot()
```

```python
df_t_avg.xs("United States")["AverageTemperature"].plot()
```

```python
df_t_median = df_t_avg["AverageTemperature"][["median"]].reset_index()
df_t_median
```

```python
df_t_median["date"] =df_t_median["year"].dt.year
df_t_median["date"]
```

```python
df_t_median.rename(columns={"median":"temperature"},inplace=True)
df_t_median
```

```python
df_t_pivot = df_t_median.pivot_table(values="temperature",index="date",columns="Country")
df_t_pivot
```

```python
df_t_pivot.boxplot()
```
