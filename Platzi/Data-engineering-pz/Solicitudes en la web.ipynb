Extracción de información de la web

```python
import bs4
soup = bs4.BeautifulSoup(response.text,"html.parser")
print(soup.title.text)
```

```python
print(soup)
```

```python
print(soup.select("meta[name=description]"))
```

```python
courses_link = soup.select(".Categories-category-item")
courses = [course["href"] for course in courses_link]
for course in courses:
    print(course)
```
